# Historia das <span style="color: blue">Redes Neurais</span>

- <span style="color: blue; font-size: 20px">1943 - Primeiro modelo neural</span>
    - Warren Mcculloch (Neurocientista), em pareceria com Walter Pitts (Lógico), desejava um modelo computacional que simulasse o cérebro humano.
    - Redes Neurais são inspiradas no cérebro humano. Sua unidade básica é o neurônio, composto por 3 partes principaiis:
        1. **Dendritos**: Recebem sianis de outros neurônios.
        2. **Corpo**: Processa os sinais.
        3. **Axônio**: Transmite o sinal para outros neurônios. Sua ativação depende da força do sinal.
    - Os neurônios são combinados em uma rede capaz de assumir diferentes padrôes de ativação. Quando ouvimos uma música, por exemplo, há um padrão de ativação. Já no caso de ouvirmos um garfo arranhando um prato, outro padrão é ativado.
    <div class="row">
        <div class="col-md-6" markdown="1">
            <img src="figures/mcculloch-pitts_model.png">
        </div>

        <div class="col-md-6" markdown="1">
            <br>
            <span style="font-weight: bold; font-size: 16px">Primeira representação artificial do neurônio:</span>
            <ol>
                <li>
                    <b>Dendrito</b>: Recebe as entradas $x_1$, $x_2$, $x_3$, ..., $x_n$.
                </li>
                <li>
                    <b>Corpo</b>: Função <b>g</b> agrega as entradas.
                </li> 
                <li>
                    <b>Axônio</b>: Função <b>f</b> decide a ativação do neurônio de acordo com a saída de <b>g</b>.
                </li>
            </ol>
            <br>
            Exemplos:
            <ul>
                <li>
                    <b>AND</b>: <b>g</b>$(x) \geqslant 2$, ou seja $x_1=1$ e $x_2=1$
                </li>
                <li>
                    <b>OR</b>: <b>g</b>$(x) \geqslant 1$, ou seja $x_1=1$ ou $x_2=1$
                </li>
            </ul>
            <img src="figures/mcculloch-pitts_model_and_or.png" style="width: 400px">
        </div>
    </div>

- <span style="color: blue; font-size: 20px">1957 - Perceptron</span>
    - Frank Rosenblatt propõe uma versão melhorada do modelo neural, associando um peso $w_i$ a cada $x_i$. Os pesos erram então <ins>otimizados</ins> de acordo com o prroblema em questão.

            <br>
            <br>
            <br>
            $y = w_0 + \sum^n_{i=1}{x_iw_i}$
        </div>
    </div>
    
    - Na prática, o modelo era um hardware que ajustava os pesos com potenciômetros.

- <span style="color: red; font-size: 20px" >1969 - Problema do XOR</span>
    - Minsky e Papert destacaram as limitações do perceptron: Perceptrons só aprendem funções lineares
        - Regressão Linear
        - Classificação linear
    - Isso deu inicio ao <span style="color: red">Inverno da I.A.</span> que durou até 1986

- <span style="color: blue; font-size: 20px">1986 - Multi-Layer Perceptron</span>
    - Geoffrey Hinton e seus colegas propõem um novo método de aprendizado: **Backpropagation**. Permitindo, assim, as redes de <ins>múltiplas camadas</ins> de perceptron aprendam funções mais complexas.
    - Na mesma época foi postulado o **Teorema da Aproximação Universal**:
    > Uma rede neural feed forward com apenas uma camada (escondida) é suficiente para **representar qualquer função**, mas a camada pode ser inviavelmente grande e pode falhar em aprender e generalizarr corretamente.
    - Infelizmente na época não era possível validarr a afirmação na prática pois os modelos eram computacionalmente custosos e exigiam um grande volume de dados.

- <span style="color: blue; font-size: 20px">2006 - Deep Learning</span>
    - Surgimento de hardwarres mais robustos (GPUs)
    - Abundância de dados

# <span style="color: blue">Perceptron</span>

$ax+by+c=0 \rightarrow w_1x_1+w_2x_2+b=0$

````python
b = 0.4
w1 = -1
w2 = 4
````

Inicialmente vamos fixar o $y$ e variar o $x$, ou seja, fixar $x_2$ e variar $x_1$, para entender o comportamento da função.
<br>
<br>
<center style="font-size: 24px">
$x_2 = -\frac{w_1x_1+b}{w_2}$
</center>

````python
x = np.linspace(-2, 4, 50)
y = -(w1 * x + b) / w2
````

````python
import matplotlib.pyplot as plt
````

````python
fig, ax = plt.subplots(figsize=(16, 8))

ax.plot(x, y)

ax.axhline(0, -2, 4, color='black', linewidth=1.5)
ax.axvline(0, -1, 1, color='black', linewidth=1.5)
plt.grid()
plt.show()
````

````python
def plotline(b, w1, w2):
    x = np.linspace(-2, 4, 50)
    y = -(w1 * x + b) / w2
    
    
    fig, ax = plt.subplots(figsize=(16, 8))

    ax.plot(x, y)

    ax.axhline(0, -2, 4, color='black', linewidth=1.5)
    ax.axvline(0, -1, 1, color='black', linewidth=1.5)
    
    plt.grid()
````

````python
def retorno(ponto):
    '''
    Calcula a função para o ponto
    '''
    
    return b + w1 * ponto[0] + w2 * ponto[1]
````

````python
p1 = (2, 0.4)
ret1 = retorno(p1)

p2 = (1, 0.6)
ret2 = retorno(p2)

p3 = (3, -0.4)
ret3 = retorno(p3)
````

````python
plotline(b, w1, w2)

plt.scatter(p1[0], p1[1], marker='o', s=100, color='green', label=f'p1 = {p1}')
plt.scatter(p2[0], p2[1], marker='o', s=100, color='orange', label=f'p2 = {p2}')
plt.scatter(p3[0], p3[1], marker='o', s=100, color='red', label=f'p3 = {p3}')

plt.legend()
plt.show()
````

Qualquer ponto sob a reta, ao jogar na equação da reta, retorna 0

````python
ret1
````

Qualquer ponto acima da reta vai ter um valor positivo

````python
ret2
````

Qualquer ponto abaxo da reta vai ter um valor negativo

````python
ret3
````

Ao calcular um ponto usando a equação da reta, temos como resposta tanto a distância em x quanto se está acima ou abaixo da reta. Por esse motivo uma reta consegue funcionar como um classficador

Em duas dimensões nosso modelo forma uma reta, como vimos anteriormente. Para um número de dimensões $d>2$, modelos lineares são chamados de **hiperplanos**, e são compostos por:
- Um peso $w_i$ para cada dimensão $x_i$.
- Um único viés $b$

De forma geral:
<center style="font-size: 25px">
    $y = b+\sum^{d}_{i=1}{w_ix_i}$
</center>

Fica evidente o motivo do <span style="color: blue">Perceptron</span> modelar retas, ele é definido por um modelo linear.

````python
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
````

````python
def plot3d(perceptron):
    w1, w2, w3 = perceptron.weight.data.numpy()[0]
    b = perceptron.bias.data.numpy()[0]
    
    X1 = np.linspace(-1, 1, 10)
    X2 = np.linspace(-1, 1, 10)
    
    X1, X2 = np.meshgrid(X1, X2)
    
    X3 = (b - w1 * X1 - w2 * X2) / w3
    
    fig = plt.figure(figsize=(6, 6))
    ax = Axes3D(fig, auto_add_to_figure=False)
    fig.add_axes(ax)
    
    ax.view_init(azim=180)
    ax.plot_surface(X1, X2, X3, cmap='plasma')
````

# <span style="color: blue">Funções de ativação</span>

De forma abstrata, um neurônio interpreta as suas entradas e libera uma **ativação** com determinada força.

Ex.:
- Um neurônio que <span style="color: green; font-weight: bold">ativa</span> quando houve uma música boa
- O mesmo neurônio fica <span style="color: red; font-weight: bold">inativo</span> quando bate o dedo na quina da parede

A ideia de quando treinamos um neurônio <span style="color: blue">Percecptron</span> é de "ensinar" a só ativar em determinados momentos. O papel da <span style="color: blue">Função de Ativação</span> é justamente definir se o neurônio vai ativar ou não e qual a força.

No primeiro código escrito para classificar um ponto tinhamos a seguinte função:
```python
def classify(ponto):
    ret = retorno(ponto)
    
    #ATIVAÇÃO
    if ret >= 0:
        return 1, 'yellow'
    else:
        return 0, 'purple'
```

````python
def plotmodel(b, w1, w2):
    x = np.linspace(-2, 4, 50)
    y = -(w1 * x + b) / w2
    
    fig, ax = plt.subplots(figsize=(16, 8))

    ax.plot(x, y)
    ax.scatter(X[:, 0], X[:, 1], c=Y)

    ax.axhline(0, -2, 4, color='black', linewidth=1.5)
    ax.axvline(0, -1, 1, color='black', linewidth=1.5)
    
    plt.xlim(-4, 4)
    plt.ylim(-3, 3)
    
    plt.grid()
````

````python
activation = nn.Sigmoid()

activation
````

````python
plotmodel(b, w1, w2)

markers = ['^', 'v', '>', '<']
colors = ['r', 'g', 'b', 'gray']

for k, idx in enumerate([17, 21, 43, 66]):
    x = torch.Tensor(X[idx])

    ret = perceptron(x)
    act = activation(ret)
    
    act_limiar = 0 if ret.data < 0 else 1

    label = 'ret = {:.3f}'.format(ret.data.numpy()[0]) + f' - limiar = {act_limiar}' + ' - act = {:.3f}'.format(act.data.numpy()[0])
    plt.plot(x[0], x[1], marker=markers[k], color=colors[k], markersize=12, label=label)
    
plt.legend()
plt.show()
````

````python
activation = nn.Tanh()

plotmodel(b, w1, w2)

markers = ['^', 'v', '>', '<']
colors = ['r', 'g', 'b', 'gray']

for k, idx in enumerate([17, 21, 43, 66]):
    x = torch.Tensor(X[idx])

    ret = perceptron(x)
    act = activation(ret)
    
    act_limiar = 0 if ret.data < 0 else 1

    label = 'ret = {:.3f}'.format(ret.data.numpy()[0]) + f' - limiar = {act_limiar}' + ' - act = {:.3f}'.format(act.data.numpy()[0])
    plt.plot(x[0], x[1], marker=markers[k], color=colors[k], markersize=12, label=label)
    
plt.legend()
plt.show()
````

# <span style="color: blue">MultiLayer Perceptron</span>

Supondo agora que desejamos separar pontos em 3 classes. Um único <span style="color: blue">Perceptron</span> não seria capaz de atender a necessidade pois não tem como criar uma única reta que separe as 3 classes.

Nesse caso, treinariamos múltiplos <span style="color: blue">Perceptrons</span> especializando cada um em um padrão.

Ex.:http://vision.stanford.edu/teaching/cs231n-demos/linear-classify/

Para problemas de classificação com múltiplas categorias, é comum usarmos a função **Softmax**. Ela transforma os scores ou a "força" da saída dos <span style="color: blue">Perceptrons</span> em uma **distribuição de probabilidades**

Ela é descrita matematicamente por: 

<span style="font-size: 30px">$S(x_i) = \frac{e^{x_i}}{\sum^{n}_{j=1}{e^{x_i}}}$</span>

Anteriormente foi comentado sobre o <span style="color: red" >Problema do XOR</span>, um dos problemas que deu início ao <span style="color: red" >Inverno da I.A.</span>. Esse problema ocorreu devido a falta de eficiência se tratando de separa conjuntos não-linearmente separáveis. A solução para esse problema veio com o <span style="color: blue">MultiLayer Perceptron</span>.

Com isso usamos camadas conhecidas como <span style="color: blue">Hidden Layers</span> que precisam ter ativações não-lineares.

Outro fator que ajudou bastante foi o <span style="color: blue">Backpropagation</span>, que insere um fluxo de backward com o cálculo do <span style="color: blue">Gradiente</span> e a <span style="color: blue">atualização</span> dos pesos.

Ex.:https://cs.stanford.edu/people/karpathy/convnetjs/demo/classify2d.html

(As características aprendidas nas camadas intermediárias são linearmente separáveis, por serem perceptrons)

### `nn.Sequential`

É um container onde se coloca múltiplos módulol. Ao realizar `forward` em um objeto `Sequential` ele aplicará sequencialmente os módulos para gerar uma saída.

````python
input_size = 2
hidden_size = 8
output_size = 1

net = nn.Sequential(
    nn.Linear(in_features=input_size, out_features=hidden_size),
    nn.ReLU(),
    nn.Linear(in_features=hidden_size, out_features=output_size)
)

net
````

### `nn.Module`

A forma mais organizada de definir modelos no PyTorch é implementando uma classe nos moldes da classe `nn.Module`. Para redes extensas, fica mais fácil a implementação e a manutenção.

Funções obrigatórias:
- `__init__()`: definição de hiperparâmetros e instância do modelo
- `forward()`: fluxo de entrada para produzir uma saída

````python
import torch

from torch import nn
````

````python
class MinhaRede(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(MinhaRede, self).__init__()
        
        self.hidden = nn.Linear(input_size, hidden_size)
        self.relu   = nn.ReLU()
        self.output = nn.Linear(hidden_size, output_size)
        
        
    def forward(self, X):
        hidden = self.relu(self.hidden(X))
        output = self.output(hidden)
        
        return output
````

````python
input_size = 2
hidden_size = 8
output_size = 1

net = MinhaRede(input_size, hidden_size, output_size)

net
````
