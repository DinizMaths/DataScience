# Relatório de Análise IV

## Seleções e Frequências

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep=';')

dados
```

### Selecione somente os imóveis classificados com tipo 'Apartamento'.

```python
selecao = dados['Tipo'] == 'Apartamento'

selecao
```

```python
df1 = dados[selecao].drop(columns=['Tipo'])
df1.index = range(df1.shape[0])

df1.head(10)
```

```python
n1 = df1.shape[0]

n1
```

### Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.

```python
tipo_imoveis = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
selecao = dados['Tipo'].isin(tipo_imoveis)

selecao
```

```python
df2 = dados[selecao]
df2.index = range(df2.shape[0])

df2.head(10)
```

```python
n2 = df2.shape[0]

n2
```

### Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.

```python
selecao = (dados['Area'] >= 60) & (dados['Area'] <= 100)

selecao
```

```python
df3 = dados[selecao]
df3.index = range(df3.shape[0])

df3.head(10)
```

```python
n3 = df3.shape[0]

n3
```

###  Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.

```python
selecao = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)

selecao
```

```python
df4 = dados[selecao]
df4.index = range(df4.shape[0])

df4.head(10)
```

```python
n4 = df4.shape[0]

n4
```

```python
print(f'Número de imóveis classificados com tipo Apartamento: \033[96m{n1}\033[0m')
print(f'Número de imóveis classificados com tipos Casa, Casa de Condomínio e Casa de Vila: \033[96m{n2}\033[0m')
print(f'Número de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites: \033[96m{n3}\033[0m')
print(f'Número de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00: \033[96m{n4}\033[0m')
```
